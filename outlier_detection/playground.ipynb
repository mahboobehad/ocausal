```python
# In the name of Allah

import numpy as np 
import pandas as pd 

dp_count = 200
df = pd.DataFrame({'pct0' : np.random.randn(dp_count) * 100,
                   'pct1' : np.random.randn(dp_count) * 100,
                   'objects': np.random.randn(dp_count) * 100})

df
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

tf_cube = plt.figure().gca(projection='3d')
tf_cube.scatter(df['pct0'], df['pct1'], df['objects'])
tf_cube.set_xlabel('pct0')
tf_cube.set_ylabel('pct1')
tf_cube.set_zlabel('objects')
plt.show()
```

```python
from numpy.linalg import inv
mu = np.array(df.mean())
cov_inv = inv(df.cov())
dp = np.array(df.iloc[1])
mahala_dist = np.sqrt((mu - dp).T @ cov_inv @ (mu - dp))
mahala_dist
```
